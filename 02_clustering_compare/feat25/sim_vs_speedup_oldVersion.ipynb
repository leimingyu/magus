```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

%matplotlib inline
```

### MC_SiglAsnOptP

```python
myTitle='MC_SiglAsnOptP'

similarity=np.array([0.388, 0.571, 0.697])

avg_speedup=np.array([0.725, 0.915,  0.735])

labels = ['interval', 'binopt', 'reduction']

# opt1

fig, ax = plt.subplots()

idx = [ 2*i+1 for i in xrange(len(labels))]

ax.plot(idx, similarity, marker='^', markersize = 13, label='Similarity')
ax.plot(idx, avg_speedup, marker='o', markersize = 13, linestyle='--', color='r', label='AvgSpeedup')

ax.set_xticks([1, 3, 5])
ax.set_xticklabels(labels)

ax.set_xlim((0, 6))


plt.xlabel(myTitle, fontsize=15, labelpad=10)  # add space between xlabel and xticklabel

leg = ax.legend(
    ncol=2, 
    scatterpoints=1,
    labelspacing=1,
    loc='upper center',
    bbox_to_anchor=(0, 1.1, 1., 0.05),
    fontsize =13)

leg.get_frame().set_alpha(1)
leg.get_frame().set_linewidth(1)
leg.get_frame().set_edgecolor("k")



# We change the fontsize of minor ticks label 
plt.tick_params(axis='both', which='major', labelsize=13)


plt.tight_layout()
plt.show()


```

```python



fig, ax = plt.subplots()
ax.scatter(similarity, avg_speedup, s=270, alpha=0.9)

ax.set_xlabel('Similarity', fontsize=15)
ax.set_ylabel('Avg Speedup', fontsize=15)
ax.set_title(myTitle)

# ax.grid(True)
# ax.yaxis.grid() # horizontal lines
# ax.xaxis.grid() # vertical lines

for i, app in enumerate(labels):
    ax.annotate(app, 
                xy=(similarity[i], avg_speedup[i]),
                xytext=(5, 30),
                textcoords='offset points', ha='center', va='center', size=13,
                arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=-0.3'))

fig.tight_layout()

plt.show()


fig.savefig(myTitle + '.pdf',  transparent = True, bbox_inches='tight')
fig.savefig(myTitle + '.png',  dpi=300, transparent = True, bbox_inches='tight')
```
