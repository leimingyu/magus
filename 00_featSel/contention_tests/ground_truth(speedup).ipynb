```python
# app name in the result_iter100

app_list = ['convfft2d', 
            'interval', 
            'reduction', 
            'scan', 
            'binopt', 
            'radixSortThrust', 
            'MCSingleAsianOptionP',
            'quasirandomGenerator', 
            'mergeSort' , 
            'transpose', 
            'sortingNetworks', 
            'matrixMul', 
            'SobolQRNG']

print len(app_list)

single_exe = {}
single_exe['MCSingleAsianOptionP'] = 494.83
single_exe['binopt'] = 25347
single_exe['scan'] = 2650.76
single_exe['radixSortThrust'] = 3228.42
single_exe['transpose'] = 743.9
single_exe['quasirandomGenerator'] = 1192.34
single_exe['sortingNetworks'] = 5819.04
single_exe['matrixMul'] = 481.79
single_exe['reduction'] = 1160.41
single_exe['SobolQRNG'] = 659.2
single_exe['interval'] = 2061.01
single_exe['convfft2d'] = 3927.42
single_exe['mergeSort'] = 873.77

print len(single_exe)
```

```python
import re
import sys

# dict of lists
from collections import defaultdict
result_dlist = defaultdict(list)

with open("test_cases/benchmark.result_iter100_modify", "r") as File:
    for line in File:
        speedup_dd = {}
        # sep line with :
        print line
        
        [part1, part2] = re.split(": \t", line)
        #print part1
        
        
        #---------------
        # extract app name
        #---------------
        [app1, app2] = re.split("_", part1)
#         sep_loc = part1.index("_")
#         app1 = part1[:sep_loc]
#         app2 = part1[sep_loc+1:]
        print app1
        print app2
        
        if app1 not in app_list or app2 not in app_list:
            print line
            print "app name not in the list"
            sys.exit(1)
            
        #---------------
        # extract app runtime
        #---------------
        
        [app1_str, app2_str]= re.split("\t", part2)
        #print app1_str, app2_str
        # extract decimals from string
        app1_runtime = re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", app1_str.split(":")[1])
#         print type(app1_runtime)
        app1_runtime = float(app1_runtime[0])
        
        app2_runtime = re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", app2_str.split(":")[1])
        app2_runtime = float(app2_runtime[0])
        
        print app1_runtime, app2_runtime
        
        app1_speedup = single_exe[app1] / app1_runtime
        app2_speedup = single_exe[app2] / app2_runtime
        
        print app1_speedup, app2_speedup

        app1and2 = app1 + "+" + app2
        #print app1and2
        
        speedup_dd[app1and2] = [app1, app1_speedup, app2, app2_speedup] 
        
        result_dlist[app1].append(speedup_dd)

```
